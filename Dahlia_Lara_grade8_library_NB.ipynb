```python
import requests

url = "https://raw.githubusercontent.com/ZoeDLara/Dahlia-Lara-grade8-library-Repo/main/books.json"
response = requests.get(url)
books_data = response.json()

print("Books loaded successfully!")
print(books_data)
```

```python
import firebase_admin
from firebase_admin import credentials, db

cred = credentials.Certificate("firebase_key.json")

if not firebase_admin._apps:
    firebase_admin.initialize_app(cred, {
        'databaseURL': 'https://dahlia-lara-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app'
    })
else:
    print("Firebase already initialized!")
```

```python
ref = db.reference("library")
ref.set(books_data)
print("Uploaded to Firebase successfully!")
```

```python
data = ref.get()
print(data)
```

```python
for book in data["books"]:
    print(book["title"])
```

```python
for book in data["books"]:
    print(book["title"], "-", book["details"]["available"])
```

```python
total = sum(book["details"]["copies"] for book in data["books"])
print("Total books in library:", total)
```
